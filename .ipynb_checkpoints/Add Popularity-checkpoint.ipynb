```python
import numpy as np
import pandas as pd
import json
from ast import literal_eval
import collections
```

```python
data = pd.read_csv('Horror Movies with Tags.csv')
data.head(n=10)
```

```python
popularity_val = []
popularity=125
for i in range(1,2501):
    popularity_val.append(popularity)
    if i%20==0:
        popularity-=1
```

```python
data['Popularity']=popularity_val
```

```python
data.head()
```

```python
#Creating a new field to club the tags and genres
Tags_Genres_list=[]

for i in range(len(data)):
    gen = data['Genres'].iloc[i]
    s = ""
    for g in gen.split(): 
        if g[-1]==',':
            g=g[:-1]
        s+=g.lower()+","
    
    tag = literal_eval(data['Tags'].iloc[i])
    for t in tag:
        if t[-1]==',':
            t=t[:-1]
        s+=t.lower()+","
    
    Tags_Genres_list.append(s)       
```

```python
#Tags_Genres_list
data['TagsGenres'] = Tags_Genres_list
```
